```python
from simalign import SentenceAligner

aligner = SentenceAligner(model="bert")
```

```python
alignments = aligner.get_word_aligns("Under the lake, probably, said Ron, squinting around at the dark, slimy walls.", "— Presupun că sub lac, zise Ron, încercând să zărească ceva pe pereţii întunecaţi, slinoşi.")
```

```python
def get_position_in_sentence(word, sentence):
    return sentence.split(" ").index(word)
```

```python
def is_match(word_en, sentence_en, word_ro, sentence_ro):
    alignments = aligner.get_word_aligns(sentence_en, sentence_ro)["mwmf"]
    return (
        (
            get_position_in_sentence(word_en, sentence_en),
            get_position_in_sentence(word_ro, sentence_ro),
        )
        in alignments,
    )
```

```python
is_match(
    "squinting",
    "Under the lake, probably, said Ron, squinting around at the dark, slimy walls.",
    "încercând",
    "— Presupun că sub lac, zise Ron, încercând să zărească ceva pe pereţii întunecaţi, slinoşi.",
)
```
